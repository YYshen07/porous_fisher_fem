```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors

np.random.seed(42)
size = 40
base = np.zeros((size, size))

linear_cmap_strong = colors.LinearSegmentedColormap.from_list("white_to_red_black", ["#ffffff", "#ff0000", "#8b0000", "#000000"])


def make_flag1():
    grid = base.copy()
    edge_value = 1.3
    grid[0, :] = edge_value
    grid[-1, :] = edge_value
    grid[:, 0] = edge_value
    grid[:, -1] = edge_value
    return grid


def make_flag2():
    grid = base.copy()
    high_value = 1.2
    grid[-1, :] = high_value
    grid[:, 0] = high_value
    grid[:, -1] = high_value
    return grid


def make_flag3():
    grid = base.copy()
    segment = 4
    gap = 4
    positions = range(0, size, segment + gap)
    red_value = 1.4
    for start in positions:
        end = min(size, start + segment)
        grid[0, start:end] = red_value
        grid[-1, start:end] = red_value
        grid[start:end, 0] = red_value
        grid[start:end, -1] = red_value
    return grid


def make_flag4():
    grid = base.copy()
    top = 0.6 + 0.8 * np.random.rand(size)
    bottom = 0.6 + 0.8 * np.random.rand(size)
    left = 0.6 + 0.8 * np.random.rand(size)
    right = 0.6 + 0.8 * np.random.rand(size)
    grid[0, :] = top
    grid[-1, :] = bottom
    grid[:, 0] = left
    grid[:, -1] = right
    return grid


def make_flag5():
    grid = base.copy()
    grid[-1, :] = 1.4
    left = np.linspace(0.5, 1.4, size)
    right = np.linspace(0.5, 1.4, size)
    grid[:, 0] = left
    grid[:, -1] = right
    return grid


def make_smooth_patches(length):
    vals = np.zeros(length)
    base_period = max(8, round(length / 6))
    base_half = max(1, round(base_period / 10))
    pos = round(base_period / 2)
    while pos <= length:
        center = np.clip(pos + np.random.randint(-2, 3), 1, length)
        half = np.clip(base_half + np.random.randint(-1, 2), 1, max(1, length // 4))
        start = max(1, center - half)
        end = min(length, center + half)
        amp = 0.5 + 1.0 * np.random.rand()
        vals[start - 1:end] = amp
        pos += base_period + np.random.randint(-2, 3)
    return vals


def make_flag6():
    grid = base.copy()
    grid[0, :] = make_smooth_patches(size)
    grid[-1, :] = make_smooth_patches(size)
    left = make_smooth_patches(size)
    right = make_smooth_patches(size)
    grid[:, 0] = np.maximum(grid[:, 0], left)
    grid[:, -1] = np.maximum(grid[:, -1], right)
    return grid

flags = [make_flag1(), make_flag2(), make_flag3(), make_flag4(), make_flag5(), make_flag6()]
for flag in flags:
    fig, ax = plt.subplots(figsize=(4, 4))
    ax.imshow(flag, cmap=linear_cmap_strong, vmin=0.0, vmax=1.5)
    ax.axis('off')
    plt.show()
```
